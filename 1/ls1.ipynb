## Задание 1
Посмотреть документацию к API GitHub, разобраться как вывести список репозиториев для конкретного пользователя, сохранить JSON-вывод в файле *.json.

```python
import requests
import json
```

```python
service = 'https://api.github.com'
user = 'mwaskom'
response = requests.get(f'{service}/users/{user}/repos')

with open('github.json', 'w', encoding='utf-8') as a:
    json.dump(response.json(), a, ensure_ascii=False)

for i in response.json():
    print(i['name'])
```
